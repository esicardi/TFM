```python
import random

class CedulaUruguaya:
    def get_validation_digit(self, ci):
        ci_str = str(ci).zfill(7)  # Asegura que 'ci' tenga 7 dígitos, rellenando con ceros a la izquierda
        weights = [2, 9, 8, 7, 6, 5, 4]  # Pesos de derecha a izquierda
        total_sum = 0
        for i in range(7):
            digit = int(ci_str[6 - i])  # Reverso para aplicar pesos
            weight = weights[i]
            total_sum += digit * weight
        remainder = total_sum % 10
        validation_digit = (10 - remainder) % 10
        return validation_digit

    def clean_ci(self, ci):
        return int(str(ci).replace("-", "").replace('.', ''))

    def validate_ci(self, ci):
        ci = self.clean_ci(ci)
        ci_str = str(ci)
        if len(ci_str) != 8:
            return False
        digits = ci_str[:-1]
        dig = int(ci_str[-1])  # El último dígito del CI
        return dig == self.get_validation_digit(int(digits))

    def random_ci(self):
        ci = random.randint(0, 9999999)  # Genera un número entre 0 y 9999999
        ci_str = f"{ci:07d}"  # Asegura que el número tenga 7 dígitos
        validation_digit = self.get_validation_digit(int(ci_str))
        return int(f"{ci_str}{validation_digit}")  # Añade el dígito verificador al final
```
